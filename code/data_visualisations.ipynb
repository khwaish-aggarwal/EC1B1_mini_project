```python
import io
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
```

```python
df_cleaned= pd.read_pickle('../data/df_cleaned.pkl')
```

```python
italy_data = df_cleaned[df_cleaned['country'] == 'Italy']
us_data = df_cleaned[df_cleaned['country'] == 'United States']

# ✅ Create a figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))  # 1 row, 2 columns

# ✅ Plot for Italy
ax1.scatter(italy_data['year'], italy_data['Real GDP'], 
            facecolor='lightskyblue', edgecolor='black', linewidth=0.5)
ax1.set_title('Real GDP of Italy over Years', fontsize=16)
ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('Real GDP', fontsize=14)
ax1.grid(axis='y', alpha=0.3)  # Faint horizontal grid lines
ax1.yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))  # Comma separators on Y-axis

# ✅ Plot for the United States
ax2.scatter(us_data['year'], us_data['Real GDP'], 
            facecolor='lightskyblue', edgecolor='black', linewidth=0.5)
ax2.set_title('Real GDP of the United States over Years', fontsize=16)
ax2.set_xlabel('Year', fontsize=14)
ax2.set_ylabel('Real GDP', fontsize=14)
ax2.grid(axis='y', alpha=0.3)  # Faint horizontal grid lines
ax2.yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))  # Comma separators on Y-axis

# ✅ Adjust spacing between subplots
plt.tight_layout()

# ✅ Display the plot
plt.show() 
```
