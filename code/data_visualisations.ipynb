```python
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
import numpy as np

# ✅ Filter data for Italy and the United States
italy_data = df_cleaned[df_cleaned['country'] == 'Italy']
us_data = df_cleaned[df_cleaned['country'] == 'United States']

# ✅ Create a figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))  # 1 row, 2 columns

# =========================
# ✅ Plot for Italy
# =========================
# Bar chart
ax1.bar(italy_data['year'], italy_data['Real GDP'], 
        color='lightskyblue', edgecolor='black', linewidth=0.5, label='Real GDP')

# Linear line of best fit (Degree 1)
z_italy_linear = np.polyfit(italy_data['year'], italy_data['Real GDP'], 1)
p_italy_linear = np.poly1d(z_italy_linear)
ax1.plot(italy_data['year'], p_italy_linear(italy_data['year']), 
         color='red', linewidth=2, label='Linear Best Fit')

# Second-degree polynomial line of best fit (Degree 2)
z_italy_quad = np.polyfit(italy_data['year'], italy_data['Real GDP'], 2)
p_italy_quad = np.poly1d(z_italy_quad)
ax1.plot(italy_data['year'], p_italy_quad(italy_data['year']), 
         color='green', linewidth=2, linestyle='--', label='Quadratic Best Fit')

# Formatting
ax1.set_title('Real GDP of Italy over Years', fontsize=16)
ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('Real GDP', fontsize=14)
ax1.grid(axis='y', alpha=0.3)
ax1.yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
ax1.legend()


# =========================
# ✅ Plot for the United States
# =========================
# Bar chart
ax2.bar(us_data['year'], us_data['Real GDP'], 
        color='lightskyblue', edgecolor='black', linewidth=0.5, label='Real GDP')

# Linear line of best fit (Degree 1)
z_us_linear = np.polyfit(us_data['year'], us_data['Real GDP'], 1)
p_us_linear = np.poly1d(z_us_linear)
ax2.plot(us_data['year'], p_us_linear(us_data['year']), 
         color='red', linewidth=2, label='Linear Best Fit')

# Second-degree polynomial line of best fit (Degree 2)
z_us_quad = np.polyfit(us_data['year'], us_data['Real GDP'], 2)
p_us_quad = np.poly1d(z_us_quad)
ax2.plot(us_data['year'], p_us_quad(us_data['year']), 
         color='green', linewidth=2, linestyle='--', label='Quadratic Best Fit')

# Formatting
ax2.set_title('Real GDP of the United States over Years', fontsize=16)
ax2.set_xlabel('Year', fontsize=14)
ax2.set_ylabel('Real GDP', fontsize=14)
ax2.grid(axis='y', alpha=0.3)
ax2.yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
ax2.legend()


# ✅ Adjust spacing between subplots
plt.tight_layout()

# ✅ Display the plot
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm

# Take the log of the relevant columns
df_cleaned['log_Real_GDP'] = np.log(df_cleaned['Real GDP'])
df_cleaned['log_Capital'] = np.log(df_cleaned['Capital'])
df_cleaned['log_Labour'] = np.log(df_cleaned['Labour'])

# Define the independent variables (log_Capital and log_Labour) and the dependent variable (log_Real_GDP)
X = df_cleaned[['log_Capital', 'log_Labour']]
y = df_cleaned['log_Real_GDP']

# Run the regression without a constant
model = sm.OLS(y, X).fit()

# Print the summary of the regression
print(model.summary())
```

```python
# Use the model to predict log_Real_GDP based on log_Capital and log_Labour
df_cleaned['modeled_log_Real_GDP'] = model.predict(X)

# Display the first few rows of the dataframe to verify the new column
print(df_cleaned.head())
```

```python
# Filter data for the United Kingdom and Italy
uk_data = df_cleaned[df_cleaned['country'] == 'United Kingdom'].copy()
italy_data = df_cleaned[df_cleaned['country'] == 'Italy'].copy()

# Check if data is empty for either country
if uk_data.empty or italy_data.empty:
    print("No data found for the United Kingdom or Italy.")
else:
    # Exponentiate the modeled log GDP to get the GDP values
    uk_data['modeled_Real_GDP'] = np.exp(uk_data['modeled_log_Real_GDP'])
    italy_data['modeled_Real_GDP'] = np.exp(italy_data['modeled_log_Real_GDP'])

    # Create the plot with two subplots side by side
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 8))  # 1 row, 2 columns

    # Plot actual and modeled Real GDP for the United Kingdom
    ax1.plot(uk_data['year'], uk_data['Real GDP'], label='Actual Real GDP', marker='o')
    ax1.plot(uk_data['year'], uk_data['modeled_Real_GDP'], label='Modeled Real GDP', marker='x')
    ax1.set_title('Actual vs Modeled Real GDP for the United Kingdom', fontsize=16)
    ax1.set_xlabel('Year', fontsize=14)
    ax1.set_ylabel('Real GDP', fontsize=14)
    ax1.yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
    ax1.grid(True, linestyle='--', alpha=0.5)
    ax1.legend()

    # Plot actual and modeled Real GDP for Italy
    ax2.plot(italy_data['year'], italy_data['Real GDP'], label='Actual Real GDP', marker='o')
    ax2.plot(italy_data['year'], italy_data['modeled_Real_GDP'], label='Modeled Real GDP', marker='x')
    ax2.set_title('Actual vs Modeled Real GDP for Italy', fontsize=16)
    ax2.set_xlabel('Year', fontsize=14)
    ax2.set_ylabel('Real GDP', fontsize=14)
    ax2.yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
    ax2.grid(True, linestyle='--', alpha=0.5)
    ax2.legend()

    # Show the plot
    plt.tight_layout()
    plt.show()
```
